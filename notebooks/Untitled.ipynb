```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({
    "Categoria": ["A", "B", "C", "D"],
    "Valor": [10, 20, 30, 40]
})

df.head()
```

```python
sns.barplot(x="Categoria", y="Valor", data=df)
plt.title("Gr√°fico de Barras - Exemplo")
plt.xlabel("Categoria")
plt.ylabel("Valor")
plt.show()
```
